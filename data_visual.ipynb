```python
#  importing  matplotlib
import  matplotlib.pyplot   as  plt
```

```python
#  we can also create two list
x1=[2,6,8,12]
y1=[3,7,12,20]

x2=[12,16,18,2]
y2=[13,17,2,12]
```

```python
plt.xlabel("Distance")  #  x axis  label
plt.ylabel("Time")      #  y 
plt.grid(c='green') #  to put  graph lines 
plt.plot(x1,y1,label="cars")
plt.plot(x2,y2,label="bikes")
plt.legend()   #  to show the lable of plot 
```

#  bar plots

```python
plt.xlabel('time')
plt.ylabel('price')
plt.bar(x1,y1,label="apple")
plt.bar(x2,y2,label="microsoft")
plt.plot(x1,y1,label="amazon",c="green")
plt.grid(c='red')
plt.legend()
```

## cricket  score

# scatter / dots  plot

```python
plt.scatter(x1,y1,marker='*',s=100)   #  marker and size  are optional 
plt.scatter(x2,y2,marker='^',s=120)
plt.plot(x1,y1)
plt.grid(c='y')
```
